### Add the email reply step, show time. None of the data is saved at this point.

```python
import tkinter as tk
from tkinter import scrolledtext
import time

def update_timer(start_time, timer_label):
    elapsed_time = time.time() - start_time
    minutes, seconds = divmod(int(elapsed_time), 60)
    timer_label.config(text=f"{minutes:02}:{seconds:02}")
    timer_label.after(1000, update_timer, start_time, timer_label)  # Schedule timer updates every 1000 milliseconds

def open_reply_window():
    # Destroy the instructions window
    root.destroy()

    # Create the reply window
    reply_window = tk.Tk()
    reply_window.title("Reply to Supervisor")
    reply_window.geometry("600x400")

    # Display the email-like message at the top
    email_message = ("Dear Team,\n\n"
                     "I hope this message finds you well. As part of our ongoing project, "
                     "I would like you to give me a brief update on your current progress. "
                     "Please include any significant achievements, challenges you are facing, "
                     "and any immediate assistance you might need from the management team.\n\n"
                     "Best regards,\n"
                     "Your Supervisor")
    message_label = tk.Label(reply_window, text=email_message, justify="left", wraplength=580)
    message_label.pack(pady=(10, 20))

    # Text box for user input
    text_box = scrolledtext.ScrolledText(reply_window, width=50, height=10)
    text_box.pack()

    # Timer label
    timer_label = tk.Label(reply_window, text="00:00", font=('Helvetica', 14))
    timer_label.pack(anchor='ne')  # Place it at the top right corner

    # Start the timer
    start_time = time.time()
    update_timer(start_time, timer_label)

    # Send button
    send_button = tk.Button(reply_window, text="Send", command=lambda: print("Response sent!"))  # Placeholder for send functionality
    send_button.pack(pady=10)

    # Start the GUI for the reply window
    reply_window.mainloop()

def open_instructions_window():
    # Create the instructions window
    root.geometry("600x400")
    root.title("Instructions")

    instruction_text = ("Please read the following instructions carefully and click 'Next' when ready to proceed:\n\n"
                        "1. Ensure you are in a quiet environment.\n"
                        "2. Keep your focus on the task.\n"
                        "3. Follow all steps sequentially.\n"
                        "4. Do not rush your responses.")
    instructions_label = tk.Label(root, text=instruction_text, justify="left", wraplength=580)
    instructions_label.pack(pady=(20, 40))

    next_button = tk.Button(root, text="Next", command=open_reply_window)
    next_button.pack(pady=10)

# Main window
root = tk.Tk()
open_instructions_window()
root.mainloop()
```

```python
import tkinter as tk
from tkinter import scrolledtext, simpledialog
import time
import csv
import os

def save_data(data, filepath):
    headers = ["ID", "Total Characters", "Total Keystrokes", "Total Backspaces", "Time Elapsed"]
    file_exists = os.path.isfile(filepath)
    with open(filepath, 'a', newline='') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=headers)
        if not file_exists:
            writer.writeheader()
        writer.writerow(data)

def open_summary_window(data, root):
    summary_window = tk.Toplevel(root)
    summary_window.title("Summary of Your Input")
    summary_window.geometry("300x200")

    summary_label = tk.Label(summary_window, text=f"ID: {data['ID']}\n"
                                                  f"Total Characters: {data['Total Characters']}\n"
                                                  f"Total Keystrokes: {data['Total Keystrokes']}\n"
                                                  f"Total Backspaces: {data['Total Backspaces']}\n"
                                                  f"Time Elapsed: {data['Time Elapsed']}",
                             justify="left", wraplength=280)
    summary_label.pack(pady=20)

    # Button to return to the beginning
    return_button = tk.Button(summary_window, text="Return to Beginning", command=lambda: [summary_window.destroy(), open_instructions_window(root)])
    return_button.pack(pady=5)

    # Button to save data and end session (does not close the window now)
    save_button = tk.Button(summary_window, text="End and Save", command=lambda: save_data(data, "/Users/ariel/Desktop/Spring24/525/GUIproject/data.csv"))
    save_button.pack(pady=5)

    def return_to_start():
        summary_window.destroy()
        open_instructions_window(root)

    def end_and_save():
        save_data(data, "/Users/ariel/Desktop/Spring24/525/GUIproject/data.csv")
        root.destroy()

    return_button = tk.Button(summary_window, text="Return to Beginning", command=return_to_start)
    return_button.pack(pady=5)

    close_button = tk.Button(summary_window, text="End and Close Window", command=end_and_save)
    close_button.pack(pady=5)

def open_reply_window(root, user_id):
    reply_window = tk.Toplevel(root)
    reply_window.title("Reply to Supervisor")
    reply_window.geometry("600x400")

    email_message = ("Dear Team,\n\n"
                     "I hope this message finds you well. As part of our ongoing project, "
                     "I would like you to give me a brief update on your current progress. "
                     "Please include any significant achievements, challenges you are facing, "
                     "and any immediate assistance you might need from the management team.\n\n"
                     "Best regards,\n"
                     "Your Supervisor")
    message_label = tk.Label(reply_window, text=email_message, justify="left", wraplength=580)
    message_label.pack(pady=(10, 20))

    text_box = scrolledtext.ScrolledText(reply_window, width=50, height=10)
    text_box.pack()

    timer_label = tk.Label(reply_window, text="00:00", font=('Helvetica', 14))
    timer_label.pack(anchor='ne')

    start_time = time.time()
    keystrokes = 0
    backspaces = 0
    stop_timer = False

    def update_timer():
        if not stop_timer:
            elapsed_time = time.time() - start_time
            minutes, seconds = divmod(int(elapsed_time), 60)
            timer_label.config(text=f"{minutes:02}:{seconds:02}")
            reply_window.after(1000, update_timer)

    update_timer()

    def key_press(event):
        nonlocal keystrokes, backspaces
        if event.keysym == 'BackSpace':
            backspaces += 1
        keystrokes += 1

    text_box.bind("<KeyPress>", key_press)

    def send_response():
        nonlocal stop_timer
        stop_timer = True
        characters = len(text_box.get("1.0", "end-1c"))
        elapsed_time = time.time() - start_time
        minutes, seconds = divmod(int(elapsed_time), 60)
        time_str = f"{minutes:02}:{seconds:02}"
        data = {
            "ID": user_id,
            "Total Characters": characters,
            "Total Keystrokes": keystrokes,
            "Total Backspaces": backspaces,
            "Time Elapsed": time_str
        }
        reply_window.destroy()
        open_summary_window(data, root)

    send_button = tk.Button(reply_window, text="Send", command=send_response)
    send_button.pack(pady=10)

def open_instructions_window(root):
    instructions_window = tk.Toplevel(root)
    instructions_window.title("Instructions")
    instructions_window.geometry("600x400")

    instruction_text = ("Please enter your ID and read the following instructions carefully:\n"
                        "1. Ensure you are in a quiet environment.\n"
                        "2. Keep your focus on the task.\n"
                        "3. Follow all steps sequentially.\n"
                        "4. Do not rush your responses.")
    instructions_label = tk.Label(instructions_window, text=instruction_text, justify="left", wraplength=580)
    instructions_label.pack(pady=(20, 10))

    user_id = simpledialog.askstring("Input", "Enter your ID number:", parent=instructions_window)
    if user_id:
        next_button = tk.Button(instructions_window, text="Next", command=lambda: [open_reply_window(root, user_id), instructions_window.destroy()])
        next_button.pack(pady=10)

def main():
    root = tk.Tk()
    root.withdraw()  # Hide the root window
    open_instructions_window(root)
    root.mainloop()

if __name__ == "__main__":
    main()
```

```python
import tkinter as tk
from tkinter import simpledialog, scrolledtext
import time
import csv
import os

def save_data(data, filepath):
    headers = ["ID", "Total Characters", "Total Keystrokes", "Total Backspaces", "Time Elapsed", "Input Text 1"]
    file_exists = os.path.isfile(filepath)
    with open(filepath, 'a', newline='') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=headers)
        if not file_exists:
            writer.writeheader()
        writer.writerow(data)

def open_summary_window(data, root):
    summary_window = tk.Toplevel(root)
    summary_window.title("Summary of Your Input")
    summary_window.geometry("300x200")
    summary_label = tk.Label(summary_window, text=f"ID: {data['ID']}\n"
                                                  f"Total Characters: {data['Total Characters']}\n"
                                                  f"Total Keystrokes: {data['Total Keystrokes']}\n"
                                                  f"Total Backspaces: {data['Total Backspaces']}\n"
                                                  f"Time Elapsed: {data['Time Elapsed']}", justify="left", wraplength=280)
    summary_label.pack(pady=20)

    return_button = tk.Button(summary_window, text="Return to Beginning", command=lambda: return_to_start(summary_window, root))
    return_button.pack(pady=5)

    close_button = tk.Button(summary_window, text="End and Save", command=lambda: end_and_save(root))
    close_button.pack(pady=5)

def return_to_start(summary_window, root):
    summary_window.destroy()
    open_instructions_window(root)

def end_and_save(root):
    # This will attempt to destroy any Toplevel windows created from root, before closing the main window.
    for widget in root.winfo_children():
        if isinstance(widget, tk.Toplevel):
            widget.destroy()
    root.quit()  # Gracefully quit the Tcl interpreter, stops the mainloop
    root.destroy()  # Destroys the root window, ensures all processes are closed
    print("All windows closed and program terminated")

def open_reply_window(root, user_id):
    reply_window = tk.Toplevel(root)
    reply_window.title("Reply to Supervisor")
    reply_window.geometry("600x400")

    email_message = ("Dear Team,\n\n"
                     "I hope this message finds you well. As part of our ongoing project, "
                     "I would like you to give me a brief update on your current progress. "
                     "Please include any significant achievements, challenges you are facing, "
                     "and any immediate assistance you might need from the management team.\n\n"
                     "Best regards,\n"
                     "Your Supervisor")
    message_label = tk.Label(reply_window, text=email_message, justify="left", wraplength=580)
    message_label.pack(pady=(10, 20))

    text_box = scrolledtext.ScrolledText(reply_window, width=50, height=10)
    text_box.pack()

    timer_label = tk.Label(reply_window, text="00:00", font=('Helvetica', 14))
    timer_label.pack(anchor='ne')

    start_time = time.time()
    keystrokes = 0
    backspaces = 0
    stop_timer = False

    def update_timer():
        if not stop_timer:
            elapsed_time = time.time() - start_time
            minutes, seconds = divmod(int(elapsed_time), 60)
            timer_label.config(text=f"{minutes:02}:{seconds:02}")
            reply_window.after(1000, update_timer)

    update_timer()

    def key_press(event):
        nonlocal keystrokes, backspaces
        if event.keysym == 'BackSpace':
            backspaces += 1
        keystrokes += 1

    text_box.bind("<KeyPress>", key_press)

    def send_response():
        stop_timer = True
        characters = len(text_box.get("1.0", "end-1c"))
        input_text = text_box.get("1.0", "end-1c")  # Get the input text from text_box
        elapsed_time = time.time() - start_time  # Time in seconds
        minutes, seconds = divmod(int(elapsed_time), 60)
        time_str = f"{minutes:02}:{seconds:02}"  # For display in GUI
        total_seconds = int(elapsed_time)  # Total seconds for CSV
        data = {
            "ID": user_id,
            "Total Characters": characters,
            "Total Keystrokes": keystrokes,
            "Total Backspaces": backspaces,
            "Time Elapsed": total_seconds,  # Save total seconds in the CSV
            "Input Text 1": input_text  # Adding the user's response to the data dictionary
        }
        save_data(data, "/Users/ariel/Desktop/Spring24/525/GUIproject/data.csv")
        print("New user input saved")
        reply_window.destroy()
        open_summary_window(data, root)

    send_button = tk.Button(reply_window, text="Send", command=send_response)
    send_button.pack(pady=10)

def open_instructions_window(root):
    instructions_window = tk.Toplevel(root)
    instructions_window.title("Instructions")
    instructions_window.geometry("600x400")

    instruction_text = ("Please enter your ID and read the following instructions carefully:\n"
                        "1. Ensure you are in a quiet environment.\n"
                        "2. Keep your focus on the task.\n"
                        "3. Follow all steps sequentially.\n"
                        "4. Do not rush your responses.")
    instructions_label = tk.Label(instructions_window, text=instruction_text, justify="left", wraplength=580)
    instructions_label.pack(pady=(20, 10))

    user_id = simpledialog.askstring("Input", "Enter your ID number:", parent=instructions_window)
    if user_id:
        next_button = tk.Button(instructions_window, text="Next", command=lambda: [instructions_window.destroy(), open_reply_window(root, user_id)])
        next_button.pack(pady=10)

def main():
    root = tk.Tk()
    root.withdraw()  # Hide the root window at start
    open_instructions_window(root)
    root.mainloop()

if __name__ == "__main__":
    main()
```
